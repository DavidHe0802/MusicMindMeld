```python
import ffmpeg
import os

def convert_mp3_to_wav(input_path, output_path):
    try:
        (
            ffmpeg
            .input(input_path)
            .output(output_path, format='wav', acodec='pcm_s16le', ac=1, ar='22050')
            .run(overwrite_output=True)
        )
        print(f"Successfully converted {input_path} to {output_path}")
    except ffmpeg.Error as e:
        print(f"Failed to convert {input_path}. Error: {str(e)}")

def batch_convert_mp3_to_wav(source_folder, output_folder):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Loop over all files in the source folder
    for file_name in os.listdir(source_folder):
        if file_name.lower().endswith('.mp3'):
            input_file = os.path.join(source_folder, file_name)
            output_file = os.path.join(output_folder, file_name[:-4] + '.wav')
            convert_mp3_to_wav(input_file, output_file)

# Example usage
source_folder = 'MIDI'
output_folder = 'out'
batch_convert_mp3_to_wav(source_folder, output_folder)
```

```python
import ffmpeg
import os

def slice_wav_to_30s_segments(input_path, output_folder, file_prefix):
    try:
        # Load the audio file to check its duration
        probe = ffmpeg.probe(input_path)
        duration = float(probe['streams'][0]['duration'])
        
        # Calculate number of full 30 second segments
        num_segments = int(duration // 30)

        for i in range(num_segments):
            segment_output_path = os.path.join(output_folder, f"{file_prefix}_segment_{i+1:03d}.wav")
            (
                ffmpeg
                .input(input_path, ss=i*30, t=30)  # Start at i*30 seconds and trim to 30 seconds
                .output(segment_output_path, format='wav', acodec='pcm_s16le', ac=1, ar='22050')
                .run(overwrite_output=True)
            )
            print(f"Successfully created {segment_output_path}")
    except ffmpeg.Error as e:
        print(f"Failed to process {input_path}. Error: {str(e)}")

def batch_slice_mp3_to_30s_wav(source_folder, output_folder):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Loop over all files in the source folder
    for file_name in os.listdir(source_folder):
        if file_name.lower().endswith('.mp3'):
            input_file = os.path.join(source_folder, file_name)
            file_prefix = file_name[:-4]  # Remove the .mp3 extension for naming segments
            slice_wav_to_30s_segments(input_file, output_folder, file_prefix)

# Example usage
source_folder = 'MIDI'
output_folder = 'JAY_CHOU'
batch_slice_mp3_to_30s_wav(source_folder, output_folder)
```

```python
import os
import librosa

def data_read(directory="JAY_CHOU"):
    data = [] # Store all data (read from music)
    labels = [] # Store corresponding labels
    srs = []
    failed_files = []

    for dirname, _, filenames in os.walk(directory):
        for filename in filenames:
            filepath = os.path.join(dirname, filename)

            # Skip non-existent files
            if not os.path.isfile(filepath):
                continue

            try:
                # Using 'kaiser_best' for high-quality resampling
                y, sr = librosa.load(filepath, res_type='kaiser_best')
                label = dirname.split('/')[-1]

                # Append them to the final data
                data.append(y)
                labels.append(label)
                srs.append(sr)

            except Exception as e:
                print(f"Failed to load {filepath}: {e}")
                failed_files.append(filepath)

    print(f"Loaded {len(data)} files with {len(failed_files)} failures.")
    return data, labels, srs, failed_files

# 
data, labels, srs, failed = data_read()
```

```python
import matplotlib.pyplot as plt
import random
def check():
    fig, axi = plt.subplots(1,1) # initialize
    ind = random.randint(0, len(labels)-1) # randomly choose one index
    times = [sample/srs[ind] for sample in range(len(data[ind]))] # get time axis
    axi.plot(times, data[ind])
    axi.set_ylabel('Normalized Amplitude')
    axi.set_xlabel('Time /s')
    axi.set(title = "one example from " + labels[ind].capitalize())

    plt.show()

check()
```
